```python
import os
import pandas as pd
import pylab as pl
import datetime
%pylab inline

assert 'DFDATA' in os.environ
    
dfdata = os.getenv("DFDATA")

#Water Resevoir Levels
#New York City water supply system reservoir levels

waterlevel = pd.read_csv(dfdata + '/zkky-n5j3/1414246141/zkky-n5j3')

waterlevel.head()
```

```python
waterlevel = waterlevel.rename(columns=lambda x: x.strip()) #remove whitespace from column names
waterlevel = waterlevel[['Pepacton Date','Pepacton Storage']]

waterlevel['Pepacton Date'] = pd.to_datetime(waterlevel['Pepacton Date']).dropna()

waterlevel.head()
```

```python
waterlevel.plot(x = 'Pepacton Date', y = 'Pepacton Storage')
plt.ylabel("Millions of Gallons")
```
