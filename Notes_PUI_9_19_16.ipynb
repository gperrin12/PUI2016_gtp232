```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
np.random.seed(229)
df = 2

def chisq_size(df, n=1):
    return np.random.chisquare(df, size = n).mean()

chisq = np.random.chisquare(df, size = 100)
chisq_mean = chisq.mean()
print(chisq_mean)

print(chisq_size(df, 100))
```

```python
df = 100
chisq_mean = [chisq_size(df, n) for n in np.linspace(10, 1000, 100)]
print(chisq_mean)
```

```python
sample_sizes=list(np.linspace(10, 1000, 100))
plt.figure()
plt.scatter(sample_sizes, chisq_mean)
```
