```python
os.chdir(PUIdata)
nyc_df = pd.DataFrame()
for y in range(94, 100):
    fname = 'zbp'+ str(y) + 'totals.zip'
    zf = zipfile.ZipFile(fname)
    df = pd.read_csv(zf.open(fname.replace('.zip','.txt')))
    df.columns = [x.lower() for x in df.columns]
    df['year'] = y + 1900
    df = df[df.iloc[:,0].isin(nyc_zip_list)]
    nyc_df = nyc_df.append(df, ignore_index=True)
```

```python
for y in range(10, 15):
    fname = 'zbp'+ str(y) + 'totals.zip'
    zf = zipfile.ZipFile(fname)
    df = pd.read_csv(zf.open(fname.replace('.zip','.txt')))
    df.columns = [x.lower() for x in df.columns]
    df['year'] = y + 2000
    df = df[df.iloc[:,0].isin(nyc_zip_list)]
    nyc_df = nyc_df.append(df, ignore_index=True)

nyc_df.head()
```

```python
#reshape nyc_df so that zips are columns, year is the index, and est is what are in the cells
nyc_df_small = nyc_df[["est", "year", "zip"]]
nyc_df_pivoted = nyc_df_small.pivot(index='zip', columns='year', values='est')
nyc_df_pivoted.head()
```

```python
nyc_df_pivoted.shape
```

```python
#drop zipcodes with missing values
for i in nyc_df_pivoted.columns:
    nyc_df_pivoted = nyc_df_pivoted[np.isfinite(nyc_df_pivoted[i])]
```

```python
#standardize each zipcode so that establishment counts for each zip have standard deviation 1 and mean 0 
for i in nyc_df_pivoted.columns:
    nyc_df_pivoted[i] = [(j-nyc_df_pivoted[i].mean()) / nyc_df_pivoted[i].std() for j in nyc_df_pivoted[i]]
```

```python
#run elbow method to determine optimal k for kmeans clustering
def elbow(data, K):
#data is your input as numpy form
#K is a list of number of clusters you would like to show.
    # Run the KMeans model and save all the results for each number of clusters
    KM = [KMeans(n_clusters=k).fit(data) for k in K]
    
    # Save the centroids for each model with a increasing k
    centroids = [k.cluster_centers_ for k in KM]

    # For each k, get the distance between the data with each center. 
    D_k = [cdist(data, cent, 'euclidean') for cent in centroids]
    
    # But we only need the distance to the nearest centroid since we only calculate dist(x,ci) for its own cluster.
    globals()['dist'] = [np.min(D,axis=1) for D in D_k]
    
    # Calculate the Average SSE.
    avgWithinSS = [sum(d)/data.shape[0] for d in dist]
    
    
    # elbow curve
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(K, avgWithinSS, 'b*-')
    plt.grid(True)
    plt.xlabel('Number of clusters')
    plt.ylabel('Average within-cluster sum of squares')
    plt.title('Elbow for KMeans clustering')
    plt.show()
    
    
    # Total with-in sum of square plot. Another way to show the result.
    wcss = [sum(d**2) for d in dist]
    tss = sum(pdist(data)**2)/data.shape[0]
    bss = tss-wcss
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(K, bss/tss*100, 'b*-')
    plt.grid(True)
    plt.xlabel('Number of clusters')
    plt.ylabel('Percentage of variance explained')
    plt.title('Elbow for KMeans clustering')
    plt.show()
```

```python
#k = 3 i think...?
elbow(nyc_df_pivoted, range(1, 20))
```

```python
#group zipcodes with kmeans method
n = 3 # number of clusters

#train the model.
km = KMeans(random_state=123, n_clusters=n)
res = km.fit(nyc_df_pivoted)
#result.
print(res.labels_)
#print res.predict(dd)
print(km.cluster_centers_)
```

```python
#add kmeans groups label
nyc_df_pivoted['kmeans_group'] = res.labels_
```

```python
#plot the data and the cluster centers
fig = plt.figure(1, figsize=(8, 6))
colors = ['red', 'blue', 'black']
val = [int(j) for j in nyc_df_pivoted.columns[0:-1]]

for i in range(len(nyc_df_pivoted)):
    plt.plot(val, nyc_df_pivoted.iloc[i,:-1], c=colors[nyc_df_pivoted.kmeans_group.iloc[i]], alpha=0.05, label='_zipcode time series')  


plt.plot(val, km.cluster_centers_[0], c=colors[0], label='group 0 cluster center')
plt.plot(val, km.cluster_centers_[1], c=colors[1], label='group 1 cluster center')
plt.plot(val, km.cluster_centers_[2], c=colors[2], label='group 2 cluster center')

plt.ylabel('standarized (by year) number of establishments')
plt.xlabel('year')
plt.xticks((np.arange(1994,2015,2)))
plt.title("Kmeans-{} Clusters, Clustering on Zipcode".format(n))
plt.legend()
plt.show()
```

The plot above shows for every year the number of establishments (standardized). Each dot represents a zipcode/year combination, and the zipcodes are clustered into three groups with the kmeans clustering method with similar numbers of establishments across years, each group represented with a different color. The darkest lines are the cluster centers for each group.

```python
#group zipcodes with gaussian mixture method
n = 3 # number of clusters

#train the model.
g = mixture.GMM(n_components=n,random_state=324)
gfit = g.fit(nyc_df_pivoted.iloc[:,0:-1])
#result.
print gfit.fit_predict(nyc_df_pivoted.iloc[:,0:-1])
#centers
print gfit.means_
```

```python
#add gaussian mixture groups label
nyc_df_pivoted['gaussian_mix_group'] = gfit.fit_predict(nyc_df_pivoted.iloc[:,0:-1])
```

```python
#plot the data and the cluster centers
fig = plt.figure(1, figsize=(8, 6))
colors = ['black', 'blue', 'red']
val = [int(j) for j in nyc_df_pivoted.columns[0:-2]]

for i in range(len(nyc_df_pivoted)):
    plt.plot(val, nyc_df_pivoted.iloc[i,:-2], c=colors[nyc_df_pivoted.gaussian_mix_group.iloc[i]], alpha=0.05, label='_zipcode time series')  


plt.plot(val, gfit.means_[0], c=colors[0], label='group 0 cluster center')
plt.plot(val, gfit.means_[1], c=colors[1], label='group 1 cluster center')
plt.plot(val, gfit.means_[2], c=colors[2], label='group 2 cluster center')

plt.ylabel('standarized (by year) number of establishments')
plt.xlabel('year')
plt.xticks((np.arange(1994,2015,2)))
plt.title("Gaussian Mixture Model-{} Clusters, Clustering on Zipcode".format(n))
plt.legend()
plt.show()
```

The plot above shows for every year the number of establishments (standardized). Each dot represents a zipcode/year combination, and the zipcodes are clustered into three groups with a gaussian mixture method with similar numbers of establishments across years, each group represented with a different color. The darkest lines are the cluster centers for each group. There are a few zipcodes who switched groups between the highest and 2nd highest groups between the kmeans and gaussian mixture methods!

## Interpreting Results

Comparing the results of the gaussian mixture model and the kmeans model, you can see from the time series graphs above that the gaussian mixture model classifies a few zipcodes that are the highest in the middle group as being in the top group. The kmeans classifies those same zipcodes in the middle group, despite one zipcode from the highest group intersecting those zipcodes in recent years. Also, the highest group for both methods exhibits a downward trend, meaning the zipcodes with the highest establishment counts have a decreasing amount of establishments over time. The middle group, however, exhibits an upward trend in the number of establishments.

Looking at the maps, we see that for both models, midtown manhattan is grouped together, likely representing the group with the highest number of establishments. The outer zipcodes of the outer boroughs, which are mostly residential, are also similarly grouped, likely representing the group with the lowest number of establishments.
